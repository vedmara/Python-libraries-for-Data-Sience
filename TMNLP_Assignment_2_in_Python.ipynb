```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import re
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

Importing Textdata. The essential cleaning of the data.

```python
file_path = '/content/drive/MyDrive/Springboard_Python/Textdata.txt'

# Open and read the file
with open(file_path, 'r') as file:
    text_data = file.read()

# Print the first few characters to verify
print(text_data[:200])  # Print first 200 characters as an example
```

```python
# Define a function to clean the text data
def clean_text(text):
    # Remove parentheses
    text = re.sub(r'\(.*?\)', '', text)
    # Remove punctuation marks
    text = re.sub(r'[^\w\s]', '', text)
    return text

# Clean the text data
text_data = clean_text(text_data)
```

```python
# Split text into words
words = text_data.split()
```

Top 20 Words Sorted by Frequency

```python
# Count word frequencies
word_freq = Counter(words)

# Get the top 20 words
top_20_words = word_freq.most_common(20)
print("Top 20 words sorted by frequency:")
for word, freq in top_20_words:
    print(f"{word}: {freq}")
```

Creating a Wordcloud

```python
# Generate the word cloud
wordcloud = WordCloud(background_color='white').generate(str(word_freq))

# Display the word cloud
plt.figure(figsize=(10, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Word ‘the’ has largest size, indicating most frequent word followed by ‘and’ and ‘to’.

Determining sentiments based on keywords present in the review text.

```python
# Count lines containing sentiments
negative_count = text_data.count('bad') + text_data.count('worst') + text_data.count('terrible')
neutral_count = text_data.count('okay') + text_data.count('fine') + text_data.count('alright')
positive_count = text_data.count('good') + text_data.count('great') + text_data.count('excellent')

print(f"Negative lines: {negative_count}")
print(f"Neutral lines: {neutral_count}")
print(f"Positive lines: {positive_count}")
```

A histogram showing the 15 most popular words by frequency.

```python
# Get top 15 words and frequencies
top_15_words = [word[0] for word in word_freq.most_common(15)]
top_15_freqs = [word[1] for word in word_freq.most_common(15)]

# Plot bar graph
plt.figure(figsize=(10, 6))
plt.bar(top_15_words, top_15_freqs, color='skyblue')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Top 15 Words by Frequency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Graph shows the frequency of the top 15 words by frequency on a horizontal bar graph. “the” is the most frequent word with frequency 82.
